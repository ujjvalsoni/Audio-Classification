```python
# Compute a mel-scaled spectrogram.

x =path + str(audio_data['Clip'][0])
y, sr = librosa.load(x)
f8 = np.mean(librosa.feature.melspectrogram(y=y, sr=sr, S=None, n_fft=2048, hop_length=512, power=2.0))
# log_f8 = librosa.power_to_db(f8, ref=np.max)
f8
```

```python
# Compute root-mean-square (RMS) energy for each frame, either from the audio samples y or from a spectrogram S.
f10 = np.mean(librosa.feature.rmse(y=y, S=None, frame_length=2048, hop_length=512, center=True, pad_mode='reflect'))
f10
```

```python
# Compute the spectral centroid.
f11 = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr, S=None, n_fft=2048, hop_length=512, freq=None))
f11
```

```python
#Compute the tempogram: local autocorrelation of the onset strength envelope
f1 = np.mean(librosa.feature.tempogram(y, sr=sr, onset_envelope=None, hop_length=512, win_length=384, center=True, window='hann'))
f1
```

```python
# Computes the tonal centroid features (tonnetz), following the method of [R8787].
f3=np.mean(librosa.feature.tonnetz(y, sr=sr, chroma=None))
f3
```

```python

# Compute a chromagram from a waveform or power spectrogram.
f5 = np.mean(librosa.feature.chroma_stft(y=y, sr=sr, S=None, n_fft=2048, hop_length=512, tuning=None))
f5
```

```python
# Compute spectral flatness.
f14 = np.mean(librosa.feature.spectral_flatness(y=y, S=None, n_fft=2048, hop_length=512, amin=1e-10, power=2.0))
f14
```

```python
# Compute roll-off frequency
f15 = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr, S=None, n_fft=2048, hop_length=512, freq=None, roll_percent=0.85))
f15
```

```python
mel_spect = []
for i in range(1,len(audio_data)):
    x1 = path + str(audio_data['Clip'][i])
    print(i)
    y1, sr1 = librosa.load(x1)
    f8 = np.mean(librosa.feature.melspectrogram(y=y1, sr=sr1, S=None, n_fft=2048, hop_length=512, power=2.0))
    mel_spect.append(f8)
```

```python
# mel_spect

norm_mel_spect =[ ]
min_mel = min(mel_spect)
deno_mel = max(mel_spect)-min(mel_spect)
    
for x in mel_spect:
    res=(x-min_mel)/deno_mel;
    norm_mel_spect.append(res)
    
```

```python
rmse = []
for i in range(1,len(audio_data)):
    x1 = path + str(audio_data['Clip'][i])
    print(i)
    y1, sr1 = librosa.load(x1)
    f10 = np.mean(librosa.feature.rmse(y=y1, S=None, frame_length=2048, hop_length=512, center=True, pad_mode='reflect'))
    rmse.append(f10)
```

```python

norm_rmse =[ ]
min_rmse = min(rmse)
deno_rmse = max(rmse)-min(rmse)
    
for x in rmse:
    res=(x-min_rmse)/deno_rmse;
    norm_rmse.append(res)
    
```

```python
centroid = []
for i in range(1,len(audio_data)):
    x1 = path + str(audio_data['Clip'][i])
    print(i)
    y1, sr1 = librosa.load(x1)
    f11 = np.mean(librosa.feature.spectral_centroid(y=y1, sr=sr1, S=None, n_fft=2048, hop_length=512, freq=None))
    centroid.append(f11)
```

```python
norm_centroid =[ ]
min_centroid = min(centroid)
deno_centroid = max(centroid)-min(centroid)
    
for x in rmse:
    res=(x-min_centroid)/deno_centroid;
    norm_centroid.append(res)
    
```
