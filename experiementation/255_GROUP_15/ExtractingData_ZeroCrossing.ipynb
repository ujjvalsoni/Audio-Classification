```python
import pandas as pd
```

```python
filename = '/Users/niki/Documents/255_final_project/ESC-50-master/esc50-human.xlsx'
xl_file = pd.ExcelFile(filename, header = True)
audio_data = xl_file.parse('Raw data')
#audio_data.columns = ['label','audio_file']
audio_data.head()
```

```python
#removed = original.replace("M", "")
audio_list = []
for name in audio_data.X_AUDIO:
    #print(name)
    new_name = name.replace(".ogg","")
    #print(new_name)
    audio_list.append(new_name)
    
audio_data["X_audio1"]=pd.DataFrame(audio_list)
audio_data.head()
```

```python
import os

path = '/Users/niki/Documents/255_final_project/ESC-50-master/audio/'
audio_files = []

for filename in os.listdir(path):
    audio_files.append(filename)
#audio_files
```

```python
keyboard = []
for j in audio_data['X_audio1']:
    for name in audio_files:
        if j in name:
            keyboard.append(name)

#print(keyboard)
audio_data["audio_files"]=pd.DataFrame(keyboard)


# audio_data.drop("X_AUDIO", axis=1, inplace=True)

audio_data.drop("X_audio1", axis=1, inplace=True)
audio_data.head()
```

```python
from __future__ import print_function
import pandas as pd
import numpy as np

# matplotlib for displaying the output
import matplotlib.pyplot as plt
import matplotlib.style as ms
ms.use('seaborn-muted')
%matplotlib inline

# and IPython.display for audio output
from IPython.display import Audio
# Librosa for audio
import librosa
# And the display module for visualization
import librosa.display
```

```python
#i=0

x =path + str(audio_data['audio_files'][0])
y, sr = librosa.load(x)
f2=librosa.feature.zero_crossing_rate(y, frame_length=2048, hop_length=512, center=True)
log_f2 = pd.DataFrame(librosa.power_to_db(f2, ref=np.max))
d4=pd.DataFrame(log_f2)
```

```python
for i in range(1,len(audio_data)):
    x1 =path + str(audio_data['audio_files'][i])
    y1, sr1 = librosa.load(x1)
    f3=librosa.feature.zero_crossing_rate(y1, frame_length=2048, hop_length=512, center=True)
    log_f3 = pd.DataFrame(librosa.power_to_db(f3, ref=np.max))
    d4=pd.concat([d4,log_f3],axis=0)
```
